### 問1

自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
6

# テストケース2
13

# テストケース3
33
```

````python
def nabeatsu(num:int) -> str:
# 3の倍数かどうかを判定（3で割った余りが0）
    multiple_of_3 = (num % 3 == 0)
    # 数値を文字列に変換し、'3'が含まれているか判定
    contains_3 = ('3' in str(num))

    # 両方の条件を満たす場合は 'piyo' を返す
    if multiple_of_3 and contains_3:
        return 'piyo'
    # 3の倍数の条件のみ満たす場合は 'hoge' を返す
    elif multiple_of_3:
        return 'hoge'
    # 3を含む条件のみ満たす場合は 'huga' を返す
    elif contains_3:
        return 'huga'
    # いずれの条件も満たさない場合は空文字（または数値）を返す
    else:
        return str(num)
````

````python
case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
def combination_sum(candidates: list, target: int) -> list:
    # dp[i] に「合計が i になる組み合わせのリスト」を格納する
    # 初期状態として、target + 1 個の空リストを持つ配列を作成する
    dp = [[] for _ in range(target + 1)]
    
    # 合計が 0 になる組み合わせは「何も選ばない」という1通り（空リスト）である
    dp[0].append([])

    # 候補の数字を1つずつ取り出し、その数字を使って作れる合計値を更新していく
    # 外側のループで数字を固定することで、[2, 3] と [3, 2] のような重複を防ぐことができる
    for num in candidates:
        # num から target までループを回す（num未満の合計にはその数字は使えないため）
        for i in range(num, target + 1):
            # 「合計が (i - num) になる組み合わせ」のそれぞれに num を追加すれば、「合計が i になる新しい組み合わせ」が作れる
            for prev_comb in dp[i - num]:
                # 新しい組み合わせを作成し、dp[i] に追加する
                dp[i].append(prev_comb + [num])

    # ターゲットの値に対応する組み合わせのリストを返す
    return dp[target]
````

````python
candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 100

print(combination_sum(candidates_1, target_1))
print(combination_sum(candidates_2, target_2))
print(combination_sum(candidates_3, target_3))
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
def format_checker(text:str) -> bool:
# 括弧の対応を確認するためのリストを用意する
    stack = []
    # 閉じ括弧に対応する開き括弧を辞書として定義する
    brackets_map = {')': '(', '}': '{', ']': '['}

    for char in text:
        # 文字が開き括弧の場合、追加する
        if char in brackets_map.values():
            stack.append(char)
        # 文字が閉じ括弧の場合の判定処理
        elif char in brackets_map.keys():
            # スタックが空（対応する開き括弧がない）または、取り出した開き括弧が対応するものでない場合は整合性が取れていないため False を返す
            if not stack or stack.pop() != brackets_map[char]:
                return False
        # 括弧以外の文字は無視して次の文字へ進む

    # すべての処理を終えた際、スタックが空であれば全ての括弧が正しく閉じられているため True である
    # スタックに残っている場合は開き括弧が多すぎるため False である
    return not stack
````

````python
case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
